```python
# EX 1

user = input("Digite seu user: ")

if user == "admin":
    passwd = input("Digite sua senha: ")
    if passwd == "1234":
        print("Voce tem acesso total")
    else:
        print("Senha incorreta")
else:
    print("User incorreto")            
```

```python
# EX 2

age = int(input("Digite sua idade: "))

if age >= 18:
    if age >= 60:
        print("Idoso")
    else:
        print("Adulto")
else:
    if age >= 12:
        print("Adolescente")
    else:
        print("Criança")
```

```python
# EX 3

nota = int(input("Digite sua nota: "))

if nota >= 6:
    if nota >= 9:
        print("Aprovado com Excelencia")
    else:
        print("Aprovado")
else:
    print("Reprovado")     
```

```python
# EX 4   

number = int(input("Digite um numero: "))

if number > 0:
    if number % 2 > 0:
        print("positivo e impar")
    else:
        print("positivo e par")    
elif number == 0:
    print("Zero")
else:
    print("negativo")
```

```python
# EX 5

value = int(input("Digite o valor: "))
vip = input("Você é vip: ")

if value >= 200:
    if vip == "sim":
        newValueVip = value - (value * 20 / 100)
        print(f"Voce ganhou um desconto de 20%, o novo valor é {newValueVip}, foi descontado {value - (value * 20 / 100)} ")
    else:
        newValue = value - (value * 10 / 100)
        print(f"Voce ganhou um desconto de 10%, o novo valor é {newValue}, foi descontado {value - (value * 10 / 100)} ")
```

```python
# EX 7

order = int(input("Digite o que voce vai querer, 1 - eggs, 2 - pancakes, 3 - wafles, 4 - frutas: "))

if order == 1:
    eggResp = input("Qual serao os acompanhamentos? ")
    print(f"Voce vai querer ovos com {eggResp}")
elif order == 2:
    print("Voce vai querer pancakes")
elif order == 3:
    print("Voce vai querer wafles")
elif order == 4:
    fruitType = input("Qual sera a fruta? ")
    fruitAcom = input("E o acompanhamento? ")
    print(f"Voce vai querer {fruitType} acompanhado de {fruitAcom}")
else:
    print("Verifique se estão no cardapio")
```
