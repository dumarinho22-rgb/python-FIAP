```python
lista = [] #define uma lista vazia
nomes = 5 #define a variavel de valor 5

for i in range(nomes): #for é um loop / i é o index(variavel aleatoria) / range é o alcance 5, pois *nomes=5*
    print(f"faltam {nomes} nomes") #contador de nomes
    nome = input("digite seu nome:")
    lista.append(nome) #append adiciona ao array (lista)
    print(lista)
    nomes -= 1 #diminui 1 no contador de nomes
```

```python
numero = 3 #definição
for i in range (5): #for é um loop / i é o index(variavel aleatoria) / range é o alcance 5 (mas vai do 0 ao 4)
    numero += i #aumenta o contador (somando)
    print(numero)
```

```python
list = ["ana", "laura", "jose", "pedro",1,2,3]
cleanlist = []

for i in list:
    if i == "ana" or i == "jose":
        cleanlist.append(i)
        print(f"Lista atualizada, a lista agora é: {cleanlist}")
    else:
        print(f"o(a) {i} foi negado")
```
