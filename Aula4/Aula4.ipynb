```python
#ex.7 Crie um programa que solicite 5 nomes e insira em uma lista
lista = [] #define uma lista vazia
nomes = 5 #define a variavel de valor 5

for i in range(nomes): #for é um loop / i é o index(variavel aleatoria) / range é o alcance 5, pois *nomes=5*
    print(f"faltam {nomes} nomes") #contador de nomes
    nome = input("digite seu nome:")
    lista.append(nome) #append adiciona ao array (lista)
    print(lista)
    nomes -= 1 #diminui 1 no contador de nomes
```

```python
#ex.8 A partir da var word = 'engenharia de software'. Se o indice da word == 'e' or 'o', continue
word = "engenharia de software"
i = 0

while i < len(word):
    if word[i] == "e":
        i += 1 
        continue
    print(word[i])
    i += 1
```

```python
#ex.9 Crie um algoritmo que solicite 3 nomes e 3 idades. Deixe a informação em uma lista no formato: nome - idade
lista = [] 
nomes = 3
idade = 3 

for i in range(nomes and idade):
    print(f"faltam {nomes} nomes:") 
    print(f"faltam também {idade} idades:")
    nome = input("digite seu nome:")
    idade = int(input("digite sua idade:"))
    lista.append(nome)
    lista.append(idade) 
    print(lista)
    nomes -= 1 
    idade -= 1
```

```python
#Ex.10 Crie uma estrutura que solicite itens de compra de forma que só seja possível sair da estrutura de repetição, 
#caso o usuário digite "SAIR". Coloque os itens em uma lista
lista = []
compras = 10

for i in range(compras):
    input("O que deseja comprar?")
    print(f"Faltam {compras}compras: ")
    compras = input("Digite o que quer comprar: ")
    lista.append(compras)
    print(lista)

if compras == "SAIR":
    print("Fim da lista")
```
