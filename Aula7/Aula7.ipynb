```python
comando = "start"
if comando == "start":
    print("Iniciando!")
elif comando == "stop":
    print("Cancelando")
else:
    print("Entrada inválida")
```

```python
#Ex 1 - Match/Case

match comando:
    case "start":
        print("Iniciando!")
    case "stop":
        print("Cancelando!")
    case _:
        print("Entrada inválida!")
```

```python
#Exercício 1 - Crie um algoritmo para solicitar temperatura em celsius e transformar para fahrenheit

def fahrenheit(temp):

    fahr = (temp * 9/5) + 32

    return fahr

print(fahrenheit(300))
```
