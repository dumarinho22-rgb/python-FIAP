```python
#Dada a equação: A^2 = B^2 = C^2, solicite ao usuário o valor de A, C e calcule o valor de B

a= float(input("Digite o valor de a:"))
c= float(input("Digite o valor de c:"))

B= (a*a * c*c)**0.5
print(f"O respectivo valor de B é {B}")
```
